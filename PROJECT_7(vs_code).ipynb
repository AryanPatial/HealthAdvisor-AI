```python
import os
import random
import json
import xml.etree.ElementTree as ET

# Define the parent directory where XML files are stored
data_dir = "/content/MedQuAD"

# List of subfolders
folders = [
    "1_CancerGov_QA", "2_GARD_QA", "3_GHR_QA", "4_MPlus_Health_Topics_QA",
    "5_NIDDK_QA", "6_NINDS_QA", "7_SeniorHealth_QA", "8_NHLBI_QA_XML", "9_CDC_QA"
]

# Dictionary to store extracted Q&A data grouped by URL
qa_data_dict = {}

# Loop through each folder and process XML files
for folder in folders:
    folder_path = os.path.join(data_dir, folder)

    if os.path.exists(folder_path):
        for file in os.listdir(folder_path):
            if file.endswith(".xml"):
                file_path = os.path.join(folder_path, file)

                try:
                    # Parse XML
                    tree = ET.parse(file_path)
                    root = tree.getroot()

                    # Extract URL
                    url = root.attrib.get("url")

                    # Extract all question-answer pairs for this URL
                    if url and "https://rarediseases.info.nih.gov/gard/" not in url:
                        if url not in qa_data_dict:
                            qa_data_dict[url] = []  # Initialize list for this URL

                        for qa_pair in root.findall(".//QAPair"):
                            question = qa_pair.find("Question").text if qa_pair.find("Question") is not None else "No Question"
                            answer = qa_pair.find("Answer").text if qa_pair.find("Answer") is not None else "No Answer"

                            # Store each Q&A pair under the respective URL
                            qa_data_dict[url].append({"question": question, "answer": answer})

                except ET.ParseError:
                    print(f"Skipping invalid XML file: {file_path}")

# Convert the dictionary into a list format for JSON storage
qa_data_list = [{"url": url, "qa_pairs": qa_pairs} for url, qa_pairs in qa_data_dict.items()]

# Randomly select 2000 Q&A pairs while maintaining the structure
sampled_data = random.sample(qa_data_list, min(2000, len(qa_data_list)))

# Save to a JSON file
output_file = "/content/filtered_qa.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(sampled_data, f, indent=4, ensure_ascii=False)

print(f"✅ Extracted {len(sampled_data)} unique URLs with Q&A pairs and saved to {output_file}.")
```

```python
import os
import json
import random
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# Define the path to the filtered_qa.json file
input_file = "/content/filtered_qa.json"
output_file = "/content/scraped_qa_content.json"

# Load the filtered QA data from JSON
try:
    with open(input_file, "r", encoding="utf-8") as f:
        qa_data = json.load(f)
except FileNotFoundError:
    print(f"Error: {input_file} not found. Please ensure the file exists.")
    exit(1)
except json.JSONDecodeError:
    print(f"Error: {input_file} is not a valid JSON file.")
    exit(1)

# Extract URLs from the loaded JSON data (assuming the structure includes "url" and "qa_pairs")
ordered_urls = [item["url"] for item in qa_data if "url" in item and "qa_pairs" in item]

# Function to extract content from static websites using BeautifulSoup
def extract_static_content(url):
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        content_tags = soup.find_all(['article', 'section', 'p'])
        extracted_text = {tag.get_text(strip=True) for tag in content_tags if tag}

        return "\n".join(extracted_text) if extracted_text else "No relevant content found."
    except requests.exceptions.RequestException as e:
        return f"Error fetching {url}: {e}"

# Function to extract content from dynamic websites using Selenium
def extract_dynamic_content(url):
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920x1080")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    try:
        driver.get(url)
        time.sleep(2)  # Minimized sleep time for efficiency
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        content_tags = soup.find_all(['article', 'section', 'p'])
        extracted_text = {tag.get_text(strip=True) for tag in content_tags if tag}
    except Exception as e:
        extracted_text = {f"Error loading {url}: {e}"}
    finally:
        driver.quit()

    return "\n".join(extracted_text) if extracted_text else "No relevant content found."

# Scrape the URLs in the original order from filtered_qa.json
scraped_data = []
for index, url in enumerate(ordered_urls, start=1):
    print(f"({index}/{len(ordered_urls)}) Scraping: {url}")

    # Determine method based on URL structure
    content = extract_dynamic_content(url) if "medlineplus.gov/genetics" in url else extract_static_content(url)

    # Preserve the QA pairs and URL from filtered_qa.json, add scraped content
    for item in qa_data:
        if item["url"] == url:
            scraped_data.append({
                "url": url,
                "qa_pairs": item["qa_pairs"],
                "content": content
            })
            break

    time.sleep(0.5)  # Reduced sleep time to avoid overwhelming servers

# Save scraped data to JSON file
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(scraped_data, f, indent=4, ensure_ascii=False)

print(f"\n✅ Scraping completed. Data saved to {output_file}.")
```

#Cleaning before RAG

```python
import pandas as pd
import json

# Load the JSON file
with open("scraped_qa_content.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# Display sample rows
df.head()
```

```python
import re

def clean_medical_text(text):
    if not isinstance(text, str) or len(text.strip()) == 0:
        return None  # Ignore empty or invalid text

    # Remove author lists (e.g., "Rodriguez M, Cavallari LH, Schmidt S.")
    text = re.sub(r"([A-Z][a-z]+(?: [A-Z]\.)?(?:, [A-Z][a-z]+(?: [A-Z]\.)?)*)", "", text)

    # Remove journal references (e.g., "Eur J Pharm Sci. 2017 Nov;109S:S9-S14.")
    text = re.sub(r"[A-Za-z]+ [A-Za-z]+\. \d{4} [A-Za-z]+;\d+.*?:\d+(-\d+)?\.", "", text)

    # Remove DOI references (e.g., "doi: 10.1016/j.ejps.2017.05.017.")
    text = re.sub(r"doi:\s*\d+\.\d+/[^\s]+", "", text)

    # Remove PubMed mentions (e.g., "Citation on PubMed")
    text = re.sub(r"\b(Citation on PubMed|Free article on PubMed Central|Epub\s*\d{4}\s*[A-Za-z]+ \d{1,2})\b", "", text, flags=re.IGNORECASE)

    # Remove special characters except for periods (.), commas (,), and basic punctuation
    text = re.sub(r"[^a-zA-Z0-9.,!?%$-]", " ", text)

    # Remove extra spaces and newlines
    text = re.sub(r"\s+", " ", text).strip()

    return text if len(text) > 50 else None  # Ignore short outputs

# Apply cleaning to your dataset
df["cleaned_text"] = df["content"].apply(clean_medical_text)

# Drop empty rows
df = df.dropna(subset=["cleaned_text"])
```

#CHUNKING THE BIG DOCUMENT BEFORE CREATING ITS EMBEDDINGS.

```python
import pandas as pd
import numpy as np

# Example function to split text into chunks of a max character length (e.g., 500 chars)
def chunk_text(text, chunk_size=500):
    words = text.split()  # Split text into words
    chunks = []
    current_chunk = []

    for word in words:
        if sum(len(w) for w in current_chunk) + len(word) + len(current_chunk) <= chunk_size:
            current_chunk.append(word)
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [word]

    if current_chunk:  # Append the last chunk
        chunks.append(" ".join(current_chunk))

    return chunks

# Apply chunking to create new rows
chunked_rows = []
for _, row in df.iterrows():
    chunks = chunk_text(row["cleaned_text"], chunk_size=500)  # Adjust chunk_size as needed

    for chunk in chunks:
        chunked_rows.append({
            "url": row["url"],
            "qa_pairs": row["qa_pairs"],  # Keep QA pairs the same for each chunk
            "content": row["content"],    # Original content (if needed)
            "cleaned_text": chunk         # Store the chunked text
        })

# Convert back to DataFrame
df_chunked = pd.DataFrame(chunked_rows)

# Check the new DataFrame
print(df_chunked.head())
```

```python
import textwrap
from langchain_community.vectorstores.faiss import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import Ollama
from langchain.embeddings import HuggingFaceEmbeddings

def initialize_rag_system(faiss_index_path, model_name="llama3.1:8b", embedding_model_name="all-MiniLM-L6-v2", k=5):

    # Load embedding model
    embedding_model = HuggingFaceEmbeddings(model_name=embedding_model_name)

    # Load FAISS vector store with embeddings
    vectorstore = FAISS.load_local(faiss_index_path, embeddings=embedding_model, allow_dangerous_deserialization=True)

    # Convert FAISS store into a retriever
    retriever = vectorstore.as_retriever(search_kwargs={"k": k})

    # Load LLM model
    llm = Ollama(model=model_name)

    # Create RAG chain
    return RetrievalQA.from_chain_type(llm=llm, retriever=retriever)

def generate_response(qa_chain, query, width=80):

    response = qa_chain.run(query)
    wrapped_response = textwrap.fill(response, width=width)
    print("\n" + wrapped_response)

# Initialize the RAG system (Only need to do this once)
faiss_index_path = "/Users/patial31/Desktop/RAG-PROJECT/faiss_index_folder"
qa_chain = initialize_rag_system(faiss_index_path)

# Example usage: Querying multiple times
query1 = "What is isolated growth hormone deficiency?"
generate_response(qa_chain, query1)
```
